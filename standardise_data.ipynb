```python
import pandas as pd
```

```python
years = ["2015", "2016", "2017", "2018", "2019"]
dfs = {}
for year in years:
    dfs[year] = pd.read_csv(f"{year}.csv").assign(Year=year).fillna(0)
    

dfs["2015"].drop(columns=["Region"], inplace=True)
dfs["2016"].drop(columns=["Region"], inplace=True)

dfs["2017"].columns = list(map(lambda x: x.replace("."," "),
                               dfs["2017"].columns))

dfs["2015"].rename(
    columns={
        "Economy (GDP per Capita)":"GDP per capita"
    },    
    inplace=True)

dfs["2016"].rename(
    columns={
        "Economy (GDP per Capita)":"GDP per capita"
    },    
    inplace=True)

dfs["2017"].rename(
    columns={
        "Economy  GDP per Capita ":"GDP per capita",
        "Health  Life Expectancy ":"Health (Life Expectancy)",
        "Trust  Government Corruption ":"Trust (Government Corruption)",
        
    },    
    inplace=True)

dfs["2018"].rename(
    columns={
        "Country or region":"Country",
        "Overall rank":"Happiness Rank",
        "Score":"Happiness Score",
        "Healthy life expectancy":"Health (Life Expectancy)",
        "Freedom to make life choices":"Freedom",
        "Perceptions of corruption":"Trust (Government Corruption)"
    },
    inplace=True)

dfs["2019"].rename(
    columns={
        "Country or region":"Country",
        "Overall rank":"Happiness Rank",
        "Score":"Happiness Score",
        "Healthy life expectancy":"Health (Life Expectancy)",
        "Freedom to make life choices":"Freedom",
        "Perceptions of corruption":"Trust (Government Corruption)"
    },
    inplace=True)
```

```python
for year,df in dfs.items():
    print(year,"\n",df.columns,"\n")
```

```python
col_set = set(dfs["2019"].columns)
full_set = set()
for year,df in dfs.items():
    cols = set(df.columns)
    full_set = full_set.union(cols)
    col_set.intersection_update(cols)
print("Interserction\n",col_set)
print("\nDifference\n",full_set.difference(col_set))
```

```python
col_list = ['Country', 'Year', 'Happiness Score', 'Happiness Rank',
           'Freedom', 'Generosity', 'GDP per capita', 'Health (Life Expectancy)', 'Trust (Government Corruption)']
set(col_list).difference(col_set)
```

```python
combined = dfs["2015"][col_list]
for year,df in dfs.items():
    if year != "2015":
        combined = combined.append(df[col_list]).reset_index(drop=True)
combined
```

```python
countries =  combined.Country.value_counts()
countries[countries < 5]
```

```python
combined.replace({
    "Taiwan Province of China":"Taiwan",
    "Hong Kong S.A.R., China":"Hong Kong",
    "Trinidad & Tobago":"Trinidad and Tobago",
    "Macedonia":"North Macedonia",
    "North Cyprus":"Northern Cyprus"
    },
    inplace=True
)
combined
```

```python
countries =  combined.Country.value_counts()
below5 = countries[countries < 5].index.to_list()
below5
# unfortunately these countries / regions dont have
# data for all 5 years and will need to be removed
```

```python
combined = combined[~combined.Country.isin(below5)]
```
